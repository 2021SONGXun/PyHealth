```python
import os
import csv
import numpy as np
data_root = r'C:\Users\zqiao\Documents\datasets\image'

x_data_root = os.path.join(data_root, 'x_data')
y_data_root = os.path.join(data_root, 'y_data')
labels = []

with open(os.path.join(y_data_root, 'dignose.csv'), 'w', encoding='utf-8', newline='') as f:
    csv_write = csv.writer(f)
    csv_write.writerow(['filename','virus','bacteria','healthy'])
    for filename in os.listdir(x_data_root):
        label = np.zeros(3)
        if 'virus' in filename:
            label[0] = 1
        elif 'bacteria' in filename:
            label[1] = 1
        else:
            label[2] = 1
        row_info = [filename] + label.tolist()
        csv_write.writerow(row_info)
        print (filename)
```
